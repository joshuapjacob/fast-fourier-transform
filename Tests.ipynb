```python
from matplotlib.image import imread
import numpy as np
import matplotlib.pyplot as plt

img = imread("images/graphics.png")
plt.figure()
plt.imshow(img)
```

```python
rgb_weights = [1, 1, 1]
grayscale_image = np.dot(img[...,:3], rgb_weights)
plt.figure()
plt.imshow(grayscale_image, cmap="gray")
grayscale_image.shape
```

```python
grayscale_image_transformed = np.fft.fft2(grayscale_image)
print(grayscale_image_transformed)
final = np.fft.ifft2(grayscale_image_transformed).real
plt.figure()
plt.imshow(final, cmap="gray")
```
